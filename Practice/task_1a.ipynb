```python
import numpy as np
import cv2 as cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('lena.jpg', 0)
m,n = img.shape
```

```python
plt.imshow(img, cmap='gray')
plt.show()
```

```python
print(m,n)
```

```python
img = cv2.resize(img, (512,512))
```

```python
print(m,n)
```

```python
f = 2
sampled_img = []
sampled_img.append(img)
```

```python
for k in range(8):
    img2 = np.zeros((m//f, n//f), dtype=np.uint8)
    if img2.shape[0] >0 and img2.shape[1] >0:
        for i in range(0, m, f):
            for j in range(0, n, f):
                try:
                    img2[i//f][j//f] = img[i][j]
                except:
                    IndexError
        sampled_img.append(img2)
    f*=2
```

```python
row, col = 2,4
fig, ax = plt.subplots(row, col, figsize=(16,8))
idx = 0
for i in range(row):
    for j in range(col):
        ax[i,j].imshow(sampled_img[idx], cmap='gray')
        h = sampled_img[idx].shape[0]
        w = sampled_img[idx].shape[1]
        ax[i,j].set_title(f'{h}x{w}')
        idx+=1
plt.tight_layout
plt.show()
```
