```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
original_image = plt.imread("Fractured Spine 746x976.tif")
original_image = cv2.resize(original_image, (512, 512))
```

```python
def convert_to_grey(image):
    grey_image = image.copy()
    if(len(image.shape) == 3):
        grey_image = np.mean(image, axis=2).astype(np.uint8)
    return grey_image
```

```python
plt.imshow(original_image, cmap='gray')
```

```python
grey_image = convert_to_grey(original_image)
```

```python
plt.imshow(grey_image, cmap='gray')
```

```python
def brightness_enhancement(image, factor, low, high):
    height, width = image.shape
    for i in range(height):
        for j in range(width):
            if(image[i][j] >= low and image[i][j] <=high):
                image[i][j] += factor
            if(image[j][j] < 0):
                image[i][j] = 0
            if(image[i][j] > 255):
                image[i][j] = 255
    return image
```

```python
low = int(input('Give the low intensity'))
high = int (input('GIve the high intensity'))
factor = int(input('Give the factor'))
```

```python
brightened_image = grey_image.copy()
brightened_image = brightness_enhancement(brightened_image, factor, low, high)
```

```python
plt.figure(figsize=(16,8))
plt.subplot(1,2,1)
plt.imshow(grey_image, cmap='gray')
plt.title('original Image')
plt.subplot(1,2,2)
plt.imshow(brightened_image, cmap='gray')
plt.title('Transformed Image')
```
