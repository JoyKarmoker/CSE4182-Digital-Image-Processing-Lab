```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
def make_grey_image(image):
    if len(image.shape) == 3:
        gray_image = np.mean(image, axis=2).astype(np.uint8)
        return gray_image
    return image
```

```python
img = cv.imread('skull.jpg')
grey_image = make_grey_image(img)
```

```python
plt.imshow(grey_image, cmap='gray')
plt.show()
```

```python
grey_image = cv.resize(grey_image, (512,512))
grey_image.shape
```

```python
sampled_image = []
height, width = grey_image.shape
sampled_image.append(grey_image.copy())
```

```python
for k in range(8):
    for i in range(height):
        for j in range(width):
            grey_image[i][j] = grey_image[i][j] >> 1
    sampled_image.append(grey_image.copy())
```

```python
plt.imshow(sampled_image[0], cmap='gray')
```
